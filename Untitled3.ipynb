```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“‚ 2. Charger les donnÃ©es (remplace par le bon lien ou chemin)
url = "/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv"
df = pd.read_csv(url)

# ðŸ‘€ 3. AperÃ§u des donnÃ©es
df.head()
```

```python
# Taille du dataset
print(df.shape)

# Infos gÃ©nÃ©rales
df.info()

# Statistiques de base
df.describe()

# VÃ©rifier les valeurs manquantes
df.isnull().sum()
```

```python
sns.barplot(x='gender', y='math score', data=df)
plt.title("Moyenne des notes en math par genre")
plt.show()
```

```python
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title("Note moyenne selon l'Ã©ducation des parents")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("CorrÃ©lation entre les notes")
plt.show()
```
